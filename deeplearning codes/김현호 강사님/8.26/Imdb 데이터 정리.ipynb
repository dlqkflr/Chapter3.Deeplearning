# 각 문장의 단어를 20개로 제한



```python
from keras import preprocessing
```

```python
preprocessing.sequence.pad_sequences([[1,2,3],[4,5,6],[1,2,3,4,5,6,7,8]], maxlen= 5)
```

```python
X_train = preprocessing.sequence.pad_sequences(X_train, maxlen= 20)
X_test = preprocessing.sequence.pad_sequences(X_test, maxlen= 20)
```

```python
print (X_train.shape)
print(X_train[0])
```

# 정리

##### X_train은 총 25,000개의 문장(Sequence)이다
##### 한문장은 20개의 단어로 이루어져 있다.
##### 각 단어의 해당하는 번호는 0~9999개로 총 1만개로 이루어져있다.
##### 단어 번호가 정수 하나이지만 실제로는 만개의 성분인 원핫 인코딩 벡터로 생각해야한다.
##### (25000, 20, 10000)

```python
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense
```

```python
model = Sequential()
model.add(Embedding(10000,8, input_length= 20))
model.add(Flatten())
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss= 'binary_crossentropy', optimizer = 'rmsprop', metrics = ['acc'])
model.fit(X_train, y_train, batch_size =32, epochs =10 , validation_split =0.2)
```
